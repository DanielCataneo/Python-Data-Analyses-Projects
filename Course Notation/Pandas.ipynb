**ABRINDO ARQUIVOS COM PANDAS**

```python
import pandas as pd

dados = pd.read_csv('/content/fifa.csv')

dados
```

```python
# Lendo os 5 primerios

dados.head()

# Lendo os 5 ultimos

dados.tail()

# Lendo com parâmetros

dados.head(6)
```

```python
# Procurando em Grid

dados.iloc[2,2] # Column 2 , Row 2
```

**UTILIZANDO FUNCAO LOC E SORTED**

```python
# Localization - Where

dados.loc[dados['Age'] == 21]
```

```python
# Sorted - Ordenar
                          #ascesding = Crescente
dados.sort_values('Age' , ascending=False)
```

```python
# Sorted com mais de um dado

dados.sort_values(['Name' , 'Age'])
```

```python
# Ordenado por quem é mais pesado -- Where com Order By

dados.sort_values('Weight')
```

**ADICIONANDO UMA COLUNA**

```python
# Nome da Coluna Nova           O que estará presente nela
dados['Total'] =  dados['Acceleration'] + dados['Agility'] + dados['Reactions']

dados
```

```python
# Mostrando somente nome e total

dados = dados[['Name' , 'Total']]
dados.sort_values('Total' , ascending=False) # Usando ascending para ser do maior para o menor
```

**EXPORTANDO OS DADOS**

```python
# Para csv
dados.to_csv('dados1.csv', index=False)
```

```python
# Para Excel

dados.to_excel('dados2.xlsx', index=False)
```

```python
# Para txt

dados.to_csv('dados3.txt', index=False, sep='\t')
```
