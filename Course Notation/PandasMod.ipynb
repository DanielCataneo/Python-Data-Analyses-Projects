**PANDAS MOD**

```python
import pandas as pd
import re
```

```python
df = pd.read_csv('/content/house.csv')

df
```

```python
df.head(5)
```

**UTILIZANDO FILTROS**

```python
filter1 = df.loc[df['Rooms'] == 3] # Where

filter1
```

```python
# Filtrando com multiplas exceÃ§Ãµes
# & and
# | or

#                   Column   Condition   and   Other Condition
filter02 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h')]

filter02
```

```python
filter03 = df.loc[(df['Rooms'] == 3 ) & (df['Type'] == 'h') & (df['Price'] <= 300000)]

filter03
```

**FILTRANDO COM REGULAR EXPRESSION**

```python
# import re para nÃ£o ter problema com case sensitive

# RE = Regular Expression
filterR1 = df.loc[df['Address'].str.contains('Turner St | Turner Rd' , flags=re.I)]

# Localization

filterR1
```

**FILTRANDO COM REGX E CONDITIONS**

```python
#                     Coluna     Se contÃ©m                     E   Outra condiÃ§Ã£o
filterR2 = df.loc[df['Address'].str.contains('^59', flags=re.I) & (df['Price'] <= 500000)]

filterR2
```

**MODIFICANDO DADOS NA DATAFRAME**

```python
#          Coluna  CondiÃ§Ã£o      Coluna a ser alterada   O que colocar no Lugar
df.loc[df['SellerG'] == 'Ronald' , ['Method' , 'Type' ]] = 'Pending'
```

**GROUP BY**
