**CRIANDO GR√ÅFICOS COM MATPLOTLIB**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

**CRIANDO O PRIMEIRO GR√ÅFICO**


```python
plt.title('Gas Prices over Time (USD)')
plt.xlabel('Years')
plt.ylabel('US Dollars')

plt.plot(gas['Year'], gas['Australia'], 'r.-' ,label='Australia')
plt.plot(gas['Year'], gas['Italy'], 'y.-',label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.-',label = 'USA')
plt.plot(gas['Year'], gas['Japan'] , 'g.-',label = 'Japan')

plt.xticks(gas['Year'][::2])
plt.legend()

plt.show()
```

**CRIANDO GR√ÅFICOS A PARTIR DE LISTAS**

```python
plt.title('Gas Prices over Time (USD)')

list_countries = ['Australia', 'USA','Italy','Mexico','France','Japan']

for country in gas:
  if country in list_countries:
   plt.plot(gas['Year'], gas[country], label = [country], marker='.')


plt.xticks(gas['Year'][::2])
plt.xlabel('Years')
plt.ylabel('US Dollars')
plt.legend()

plt.savefig('gas_prices.pdf')

plt.show()

```
