```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp_vendas = {'Temperature': [37, 40, 49, 61, 72, 79, 83, 81, 75, 64, 53, 42],
               'Revenue': [292, 228, 324, 376, 440, 496, 536, 556, 496, 412, 324, 320]}

df = pd.DataFrame(temp_vendas)

df = df.sort_values(['Temperature'], ascending= False)
df
```

```python
plt.title('Ice Cream Truck')
plt.xlabel('Temperature (Fahrenheit)')
plt.ylabel('Sales (US$)')

plt.plot(df['Temperature'] , df['Revenue'] , 'b.--', label='US$')
plt.legend()


plt.show()
```
